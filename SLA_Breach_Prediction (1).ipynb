```python
# 1. Import libraries
import pandas as pd

# 2. Create dataset
data = {
    "Tickets":[10,20,30,40,50,60,70,80],
    "Priority":[1,1,2,2,3,3,3,2],
    "Hour":[9,10,11,12,13,14,15,16],
    "Breach":[0,0,0,1,0,1,1,1]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# 3. Basic EDA
print("\nSummary:")
print(df.describe())

print("\nCorrelation:")
print(df.corr())

# 4. Feature Engineering
df["Load"] = df["Tickets"] / df["Priority"]
df["Peak_Hour"] = df["Hour"].apply(lambda x: 1 if x >= 12 else 0)

print("\nData after feature engineering:")
print(df)

# 5. Define features and target
X = df[["Tickets","Priority","Hour","Load","Peak_Hour"]]
y = df["Breach"]

# 6. Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 7. Train Model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

# 8. Predictions
y_pred = model.predict(X_test)

# 9. Evaluation
from sklearn.metrics import accuracy_score, confusion_matrix

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# 10. New Prediction
new_data = pd.DataFrame(
    [[50,3,14,50/3,1]],
    columns=["Tickets","Priority","Hour","Load","Peak_Hour"]
)

prediction = model.predict(new_data)[0]

print("\nPrediction for new ticket:", prediction)

if prediction == 1:
    print("Result: SLA Breach Likely")
else:
    print("Result: No Breach")
```
